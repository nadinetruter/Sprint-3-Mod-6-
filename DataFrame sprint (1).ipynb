```python
import pandas as pd

product_dict = {'Name':['Chips','Cooldrinks','Chocolates','Pies','Fruit','Cupcakes','Veggies'],
             'Product1':['Simba','Coke','Cadbury','PepperSteak','Pear','Vanilla','Spinach'],
             'Product2':['Lays','Fanta','Tex','Chicken','Apple','Chocolate','Cabbage'],
             'Product3':['None','None','None','None','Orange','None','None']}

frame = pd.DataFrame(product_dict,index=[1,2,3,4,5,6,7])

#selecting first row
selection = ['Name','Product1','Product2','Product3']

#link selection to frame
frame = frame[selection]

#using styling functions for tables. 
frame.style.set_table_styles(
[{'selector': 'th', #header
  'props': [('background', 'orange'), 
            ('color', 'white'),
            ('font-family', 'Arial')]}, 

 {'selector': 'tr:nth-of-type(odd)', #odd rows
  'props': [('background', 'baby grey')]}, 
 
 {'selector': 'tr:nth-of-type(even)',#even rows
  'props': [('background', 'white')]},
 
  {'selector': 'tr:hover', #hover over rows
  'props': [('background-color', 'yellow')]},
 
  {'selector': 'td', #table data
  'props': [('font-family', 'Arial')]}

])
```
